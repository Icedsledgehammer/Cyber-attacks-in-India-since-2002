```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
def weight(input, output):
  mean_x = np.mean(input)
  mean_y = np.mean(output)
  theta_1 = (np.sum((input - mean_x) * (output - mean_y))) / (np.sum((input - mean_x)**2))
  
  theta_0 = mean_y - theta_1 * mean_x
  return [theta_0, theta_1]
```

```python
X_train, X_test, y_train, y_test = train_test_split(a, b, test_size = 0.2, random_state = 42)
 
# Create and train the linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error:", mse)
print("R-squared (R2) Score:", r2)
```

```python
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF, WhiteKernel, Matern
```

```python
X_train, X_test, y_train, y_test = train_test_split(a, b, test_size = 0.2, random_state = 42)

# Define the RBF kernel with a white noise term
length_scale = 1.0
nu = 1.5  # Adjust the nu parameter as needed
kernel = 1.0 * RBF(length_scale = length_scale) + 1.0 * WhiteKernel()

# Create and train the Gaussian Process Regression model
model = GaussianProcessRegressor(kernel = kernel, n_restarts_optimizer = 10, random_state = 42)
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred, y_std = model.predict(X_test, return_std = True)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error:", mse)
print("R-squared (R2) Score:", r2)

# Plot the results (optional)
plt.figure(figsize = (10, 6))
plt.scatter(X_train.index, y_train, label = 'Training Data')
plt.plot(X_test.index, y_pred, color = 'r', label = 'Predicted Mean')
plt.fill_between(X_test.index, y_pred - 1.96 * y_std, y_pred + 1.96 * y_std, color = 'b', alpha = 0.2, label = '95% Confidence Interval')
plt.xlabel('States')
plt.xticks(rotation = 90)
plt.ylabel('Attacks in numbers')
plt.title('Gaussian Process Regression with RBF kernel')
plt.legend()
plt.show()
```
